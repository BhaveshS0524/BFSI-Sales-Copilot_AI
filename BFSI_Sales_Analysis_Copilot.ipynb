```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Clean Data Loading
try:
    df = pd.read_csv('bank.csv')
    # Set a professional style for your portfolio
    sns.set_theme(style="whitegrid")
    print("✅ Data Loaded. Generating Manual Insights...")
except FileNotFoundError:
    print("❌ bank.csv not found. Please upload the file to Colab.")

def generate_bfsi_dashboard():
    # Create a layout with 2 rows and 2 columns
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('BFSI Sales & Customer Deep-Dive', fontsize=20)

    # Chart 1: Deposit Success Rate (Target Variable)
    sns.countplot(x='deposit', data=df, ax=axes[0, 0], palette='viridis')
    axes[0, 0].set_title('Overall Deposit Success (Yes/No)')

    # Chart 2: Success by Job Category
    job_order = df['job'].value_counts().index
    sns.countplot(y='job', hue='deposit', data=df, order=job_order, ax=axes[0, 1])
    axes[0, 1].set_title('Success Rate by Profession')

    # Chart 3: Age Distribution vs Success
    sns.histplot(x='age', hue='deposit', data=df, kde=True, ax=axes[1, 0], element="step")
    axes[1, 0].set_title('Age Demographics of Subscribers')

    # Chart 4: Balance vs Loan Status
    sns.boxplot(x='housing', y='balance', data=df, ax=axes[1, 1], showfliers=False)
    axes[1, 1].set_title('Balance vs. Housing Loan Status')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Run the manual dashboard
generate_bfsi_dashboard()
```
